```python
from comet import download_model, load_from_checkpoint

model_path = download_model("Unbabel/unite-mup")
model = load_from_checkpoint(model_path)
data = [
  {
    "src": "твоя моя давать задания",
    "mt": "Ты должен дать мне свои задания",
    "ref": "твоя моя давать задания"
  },
  {
    "src": "твоя моя давать задания",
    "mt": "Я даю тебе свои задания",
    "ref": "твоя моя давать задания"
  },
  {
    "src": "твоя моя давать задания",
    "mt": "Я даю тебе мои задания",
    "ref": "твоя моя давать задания"
  },
  {
    "src": "твоя моя давать задания",
    "mt": "Я должен дать тебе мои задания",
    "ref": "твоя моя давать задания"
  },
  {
    "src": "твоя моя давать задания",
    "mt": "Ты должен дать мне мои задания",
    "ref": "твоя моя давать задания"
  }
]

# param batch_size: amount of examples that will be processed from [data]
# param gpus: amount of gpus that would be used for the processing
#              1 was set for google colaboratory
model_output = model.predict(data, batch_size=5, gpus=1)

# the whole output
print("==============================")
print(model_output)

# [scores]
print("==============================")
print(model_output.scores)


```
